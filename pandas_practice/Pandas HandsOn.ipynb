# Task 1

```python
df = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/covid_19_data.csv")
```

```python
df.head(10)
```

```python
df.dtypes
```

```python
df["ObservationDate"] = pd.to_datetime(df["ObservationDate"])
df["Last Update"] = pd.to_datetime(df["Last Update"])
df
```

```python
df.dtypes
```

# Task 2

```python
updated_data = df[df["Last Update"] == pd.to_datetime("2020-06-13 03:33:14")]
updated_data
```

```python
updated_data.count()
```

# Task 3

```python
top_10 = updated_data.sort_values(by="Confirmed",ascending=False).head(10).loc[:,['Country/Region', 'Confirmed', 'Deaths', 'Recovered']]
top_10
```

```python
for item in top_10["Country/Region"].unique():
    print(item)
```

```python
top_10.groupby(by="Country/Region").sum().sort_values(by="Confirmed").head(1)
```

# Task 4

```python
top_10g = top_10.groupby("Country/Region").sum()
top_10g['Recovered_percentage'] = (top_10g["Recovered"] / top_10g["Confirmed"]) * 100
top_10g['Deaths_percentage'] =  (top_10g["Deaths"] / top_10g["Confirmed"]) * 100
```

```python
top_10g.sort_values(by="Recovered_percentage",ascending = False).head(3)
```
